```python
from nltk.stem import PorterStemmer
```

```python
words = [
    'eat','eats','eating','eaten',
    'run','runs','running','ran',
    'write','writes','writing','written',
    'study','studies','studying','studied',
    'create','creates','creating','created','creation',
    'analyze','analyzing','analyzed','analysis',
    'develop','develops','developing','developed','development',
    'play','plays','playing','played',
    'move','moves','moving','moved','movement',
    'act','acts','acting','action','active',
    'connect','connects','connecting','connected','connection',
    'secure','secures','securing','secured','security'
]
```

```python
stemming = PorterStemmer()
for word in words:
    print(word+'--->'+stemming.stem(word))
```

```python
from nltk.stem import RegexpStemmer
RegStemmer = RegexpStemmer('ing$|s$', min =4)
RegStemmer.stem('eating')
```

```python
RegStemmer.stem('ingeating')
```

```python
from nltk.stem import SnowballStemmer
```

```python
sbs=SnowballStemmer('english')
for word in words:
      print(word+'--->'+sbs.stem(word))
```
